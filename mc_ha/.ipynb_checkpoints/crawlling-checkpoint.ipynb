```python
import requests
# from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
url = f'http://www.yes24.com/Product/Search?domain=ALL&query={input()}'
req = requests.get(url)
html = req.text
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
a = soup.find_all('div','item_info')
```

```python
a
```

```python
for i in range(10):
    name = a[i].select_one('div.info_row.info_name > a.gd_name').text
    print(name)
```
